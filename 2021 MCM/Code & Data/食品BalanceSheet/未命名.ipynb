```python
import pandas as pd 
import numpy as np 
from IPython.display import display

data = pd.read_csv('FoodBalanceSheets_E_All_Data_NOFLAG.csv')
data = data[data['Element']=='Food supply (kcal/capita/day)']
data = data.drop('Element',axis=1)
data = data[data.Item.isin(['Vegetables','Meat','Eggs','Maize and products'])]
data
```

```python
all_data = []
for row in data.iterrows():
    for i in range(2,len(row[1])):
        one_row = []
        one_row.append(row[1][0])
        one_row.append(row[1][1])
        year = 2012 + i 
        one_row.append(year)
        one_row.append(row[1][i])
        all_data.append(one_row)
all_data = pd.DataFrame(all_data,columns=['Country','Item','Year','Value'])
all_data
```

```python
all_data.drop_duplicates(inplace=True)
```

```python
result = pd.DataFrame()
count = 0
for item in all_data['Item'].unique():
    one_item = all_data[all_data.Item == item]
    one_item.set_index(['Country','Year'],inplace=True)
    one_item.drop('Item',axis=1,inplace=True)
    one_item.columns = [item]
#     display(one_item)
    if count ==0 :
        
        result = one_item
    else:
        result =pd.merge(result,one_item,how='left',
                         left_index=True,right_index=True)
    count += 1
    print(count)
result
```

```python
result.to_csv('balance.csv')
```
