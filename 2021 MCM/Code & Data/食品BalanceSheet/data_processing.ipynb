```python
import os
import cv2
import numpy as np 
import pandas as pd
from IPython.display import display
path = r'D:\PyCharmproject\American\E\Raw_Data\raw_data'
files_name = os.listdir(path)
files_name
```

```python
data = pd.read_excel('./model_data.xlsx')
data = data[['Area','Year']]
all_data = data.set_index(['Area','Year'])
# all_data = data.drop(['Export Value','Import Value'],axis=1)
all_data
```

```python
for file in files_name:
    file_path = path + "\\" + file
    data = pd.read_csv(file_path)
    data.set_index(['Area','Year'],inplace=True)
    all_data = pd.merge(all_data, data, left_index=True, right_index=True, how='left')
all_data = all_data.reset_index()
all_data
```

```python
new = pd.DataFrame()

# using the area mean
for area in all_data['Area'].unique():
    df = all_data[all_data['Area']==area]
    df.fillna(method = 'backfill',inplace=True)
    new = pd.concat([new,df],axis=0)
all_data = new

# using the area mean
new = pd.DataFrame()
for area in all_data['Area'].unique():
    df = all_data[all_data['Area']==area]
    df.fillna(method = 'pad',inplace=True)
    new = pd.concat([new,df],axis=0)
all_data = new

# using this year mean
new = pd.DataFrame()
for year in all_data['Year'].unique():
    print(year)
    df = all_data[all_data['Year']==year]
    for column in list(df.columns[df.isnull().sum() > 0]):
        mean_val = df[column].mean()
        df[column].fillna(mean_val, inplace=True)
    new = pd.concat([new,df],axis=0)
all_data = new
all_data = all_data.set_index(['Area','Year'])
all_data
```

```python
all_data.to_csv('model_data.csv')
```
