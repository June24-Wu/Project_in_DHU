```python
import pandas as pd 
import numpy as np 
from IPython.display import display
data = pd.read_csv('消费者价格指数old.csv',encoding='unicode_escape')
data
```

```python
one_data = data.groupby(['Area','Item','Year']).mean()
data = one_data.reset_index()
data
```

```python
all_ = pd.DataFrame()
for i in data['Item'].unique():
    one_data = data[data['Item']==i]
    one_data.drop('Item',axis=1,inplace=True)
    one_data.columns = ['Area','Year',i]
    one_data.set_index(['Area','Year'],inplace=True)
    display(one_data)
    all_ = pd.concat([all_,one_data],axis=1)
all_.to_csv('消费者价格指数.csv')
```
