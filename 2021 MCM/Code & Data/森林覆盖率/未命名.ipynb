```python
import os
from IPython.display import display
import pandas as pd 
import numpy as np 
data = pd.read_csv('森林覆盖率raw.csv')
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[1:]
year_list
```

```python
all_data = []
for area in data['Country Name'].unique():
    print(area)
    for i in year_list:
        li = []
        li.append(area)
        li.append(i)
        li.append(data[data['Country Name']==area][i].values[0])
        all_data.append(li)
all_data = pd.DataFrame(all_data)
all_data.columns = ['Area','Year','Forest cover rate']
all_data
```

```python
all_data.fillna( method ='ffill', inplace = True)
all_data
```

```python
all_data.to_csv('森林覆盖率.csv')
```
