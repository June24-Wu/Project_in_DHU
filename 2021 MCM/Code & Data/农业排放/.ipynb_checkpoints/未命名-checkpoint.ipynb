```python
import pandas as pd 
import numpy as np
from IPython.display import display
data = pd.read_csv('Emissions_Agriculture_Energy_E_All_Data_NOFLAG.csv',encoding='unicode_escape')
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[34:]
year_list
```

```python
all_data = []
for location in data['Area'].unique():
    print(location)
#     for item in data['Item'].unique():
    for element in data['Element'].unique():
        one_row = data[
            (data['Area'] == location) &
            (data['Element'] == element) 
        ]
        for year in year_list:
            one_data = []
            one_data.append(location)
            one_data.append(element)
            one_data.append(year)
            try:
#                     value = one_row[year].values[0]
#                     if np.isnan(value) == False:
                one_data.append(one_row[year].values[0])
            except:
                 one_data.append(np.nan)
            all_data.append(one_data)
all_data = pd.DataFrame(all_data)
all_data.columns = ['Area','Element','Year','Value']
all_data
```

```python
all_data
```

```python
all_data1 = all_data
all_data1.drop('Element',axis=1,inplace=True)
all_data1.columns = ['Area','Item','Year','Area harvested(ha)']
all_data1.set_index(['Area','Item','Year'])
all_data1
```

```python
all_data.to_csv('result.csv')
```
