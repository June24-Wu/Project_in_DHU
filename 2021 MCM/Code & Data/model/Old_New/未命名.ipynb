```python
import pandas as pd 
import numpy as np 
from IPython.display import display
```

```python
data = pd.read_csv('topsis_old.csv')
# data = data[['Country','Year','Score']]
# data = data.set_index(['Country','Year'])
# Result['排序'] = Result.rank(ascending=False)['综合得分指数']
result = pd.DataFrame()
for year in data.Year.unique():
    one_year_data  = data[data.Year == year]
    one_year_data['Rank'] = one_year_data.rank(ascending=False)['Score']
    result = pd.concat([result,one_year_data],axis=0)
Result = result.set_index(['Country','Year'])
Result
```

```python
data = pd.read_csv('topsis_new.csv')
# data = data[['Country','Year','Score']]
# data = data.set_index(['Country','Year'])
# Result['排序'] = Result.rank(ascending=False)['综合得分指数']
result = pd.DataFrame()
for year in data.Year.unique():
    one_year_data  = data[data.Year == year]
    one_year_data['Rank'] = one_year_data.rank(ascending=False)['Score']
    result = pd.concat([result,one_year_data],axis=0)
result = result.set_index(['Country','Year'])
Result = pd.concat([Result,result],axis=1)
Result
```
