```python
import pandas as pd 
import numpy as np 
data = pd.read_csv('Population_E_All_Data_NOFLAG.csv',encoding='unicode_escape')
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[53:]
year_list
```

```python
from IPython.display import display
all_data = []
for location in data['Area'].unique():
    print(location)
    for item in data['Element'].unique():
        one_row = data[
            (data['Area'] == location) &
            (data['Element'] == item) 
        ]
        for year in year_list:
            one_data = []
            one_data.append(location)
            one_data.append(item)
            one_data.append(year)
            try:
                one_data.append(one_row[year].values[0])
            
            except:
                 one_data.append(np.nan)
            all_data.append(one_data)
all_data = pd.DataFrame(all_data)
all_data.columns = ['Area','Item','Year','Value']
all_data
```

```python
all_data['Value'] = all_data['Value'] * 1000
all_data
```

```python
all_data.to_csv('result.csv')
```
