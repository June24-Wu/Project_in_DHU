```python
import pandas as pd 
data = pd.read_csv('Population_E_All_Data_NOFLAG.csv',encoding='unicode_escape')
data
```

```python
year_list = data.columns.tolist()
```

```python
from IPython.display import display
all_data = []
year_list = data.columns.tolist()
year_list = year_list[4:]
year_list
for location in data['Area'].unique():
    print(location)
    for item in data['Item'].unique():
        for month in data['Months'].unique():
            one_row = data[
                (data['Area'] == location) &
                (data['Item'] == item) &
                (data['Months'] == month) 
            ]
            for year in year_list:
                try:
                    one_data = []
                    one_data.append(location)
                    one_data.append(item)
                    one_data.append(year)
                    one_data.append(month)
                    one_data.append(one_row[year].values[0])
                    all_data.append(one_data)
                except:
                    continue
all_data = pd.DataFrame(all_data)
all_data.columns = ['Area','Item','Year','Month','Value']
all_data
```

```python
all_data.to_csv('result.csv')
```
