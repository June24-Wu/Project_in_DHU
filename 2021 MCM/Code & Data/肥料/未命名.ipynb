```python
import pandas as pd 
import numpy as np 
data = pd.read_csv('Environment_Fertilizers_E_All_Data_NOFLAG.csv',encoding='unicode_escape')
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[42:]
year_list
```

```python
data = data.groupby(['Area']).sum()
data
```

```python
data.reset_index(inplace=True)
from IPython.display import display
all_data = []
for location in data['Area'].unique():
    print(location)
    one_row = data[
        (data['Area'] == location) 
    ]
    for year in year_list:
        one_data = []
        one_data.append(location)
        one_data.append(year)
        one_data.append(one_row[year].values[0])
        all_data.append(one_data)
all_data = pd.DataFrame(all_data)
all_data.columns = ['Area','Year','Nutrient']
all_data
```

```python
all_data.to_csv('result.csv')
```
