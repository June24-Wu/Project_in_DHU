```python
import pandas as pd 
import numpy as np 
from IPython.display import display
data= pd.read_excel('data.xlsx')
data
```

```python
all_result = []
for country in data['Country Name'].unique():
    print(country)
    one_country_data = data[data['Country Name'] == country]
    for i in range(1,len(one_country_data.columns)):
        one = []
        one.append(country)
        year = i+1959
        one.append(year)
        one.append(one_country_data.iloc[0,i])
        all_result.append(one)
all_result = pd.DataFrame(all_result,columns=['Country','Year','Value'])
all_result
        
        
```

```python
all_result.to_csv('result.csv')
```
