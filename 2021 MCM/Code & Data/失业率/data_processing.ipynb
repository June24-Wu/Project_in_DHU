```python
import pandas as pd 
import numpy as np 
from IPython.display import display
data = pd.read_csv('失业率old.csv')
data
```

```python
year_list = data.columns.tolist()[1:]
year_list
```

```python
all_data = []
for area in data['Country Name'].unique():
    print(area)
    one_data = data[data['Country Name']==area]
    for year in year_list:
        one = []
        one.append(area)
        one.append(year)
        one.append(one_data[year].values[0])
        all_data.append(one)
all_data = pd.DataFrame(all_data)
all_data.columns = ['Area','Year','Unemployment_rate']
all_data
#     one_data.columns = ['Area','Year',i]
all_data.set_index(['Area','Year'],inplace=True)
#     display(one_data)
#     all_ = pd.concat([all_,one_data],axis=1)
all_data.to_csv('失业率.csv')
```
