```python
import os
from IPython.display import display
import pandas as pd 
import numpy as np 
path = r'D:\PyCharmproject\American\E\人均耕地面积\raw_data'
files_name = os.listdir(path)
original = pd.DataFrame()
for files in files_name:
    file  = path + "\\" + files
    data = pd.read_excel(file)
    data.drop('序号',axis = 1 , inplace=True)
    data.set_index('时间',inplace=True)
    original  = pd.concat([original,data],axis=1)
data = original
data.reset_index(inplace=True)
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[1:]
year_list
```

```python
all_data = []
for location in data['时间'].unique():
    print(location)
#     for item in data['Element'].unique():
    one_row = data[
        (data['时间'] == location)
    ]
    for year in year_list:
        one_data = []
        one_data.append(location)
#         one_data.append(item)
        one_data.append(year)
        try:
            one_data.append(one_row[year].values[0])

        except:
             one_data.append(np.nan)
        all_data.append(one_data)
all_data = pd.DataFrame(all_data)
all_data.columns = ['时间','地点','人均耕地面积']
all_data
```

```python
all_data.to_excel('result.xlsx')
```
