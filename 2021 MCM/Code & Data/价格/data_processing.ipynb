```python
import pandas as pd 
import numpy as np
from IPython.display import display
data = pd.read_csv('Prices_E_All_Data_NOFLAG.csv',encoding='unicode_escape')
data = data[(data['Element']=='Producer Price (LCU/tonne)')]
data = data[~(data['Months'] == 'Annual value')]
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[14:]
year_list
```

```python
one_data = data[(data['Area'] == 'Afghanistan')][year_list]
one_data = pd.DataFrame(one_data.mean(),columns=['Producer Price Anuual Value'])     
one_data
# one_country_data  = pd.concat([one_country_data,one_data],axis=1)
```

```python
all_data = pd.DataFrame()
for location in data['Area'].unique():
    print(location)
    one_country_data = pd.DataFrame()
    one_data = data[(data['Area'] == location)][year_list]
    one_data = pd.DataFrame(one_data.mean(),columns=['Producer Price Anuual Value'])   
    one_data.reset_index(inplace=True)
    one_data['Area'] = location
    all_data  = pd.concat([all_data,one_data],axis=0)
all_data
```

```python
all_data.to_csv('result.csv')
```
