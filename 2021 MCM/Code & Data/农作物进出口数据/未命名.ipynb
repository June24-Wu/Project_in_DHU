```python
import pandas as pd 
import numpy as np
from IPython.display import display
data = pd.read_csv('Trade_Crops_Livestock_E_All_Data_NOFLAG.csv',encoding='unicode_escape')
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[43:]
year_list
```

```python
elements = ['Export Value','Import Value']
one_country_data = pd.DataFrame()
for element in elements:
    one_data = data[(data['Area'] == 'Afghanistan')&(data['Element']==element)][year_list]
    one_data = pd.DataFrame(one_data.sum(),columns=[element])     
    one_country_data  = pd.concat([one_country_data,one_data],axis=1)
one_country_data.reset_index(inplace=True)
one_country_data['Area'] = 'Afghanistan'
one_country_data
```

```python
elements = ['Export Value','Import Value']
all_data = pd.DataFrame()
for location in data['Area'].unique():
    print(location)
    one_country_data = pd.DataFrame()
    for element in elements:
        one_data = data[(data['Area'] == location)&(data['Element']==element)][year_list]
        one_data = pd.DataFrame(one_data.sum(),columns=[element])     
        one_country_data  = pd.concat([one_country_data,one_data],axis=1)
    one_country_data.reset_index(inplace=True)
    one_country_data['Area'] = location
    all_data  = pd.concat([all_data,one_country_data],axis=0)
all_data
```

```python
all_data.to_csv('result.csv')
```
