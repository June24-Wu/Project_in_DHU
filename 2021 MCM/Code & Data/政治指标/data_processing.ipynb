```python
import pandas as pd 
import numpy as np 
from IPython.display import display
data = pd.read_csv('WGIData.csv')
data
```

```python
year_list = data.columns.tolist()
year_list = year_list[4:]
year_list
```

```python
from IPython.display import display
all_data = []

for location in data['Area'].unique():
    print(location)
    for item in data['Item'].unique():
        one_row = data[
            (data['Area'] == location) &
            (data['Item'] == item) 
        ]
        for year in year_list:
            one_data = []
            one_data.append(location)
            one_data.append(item)
            one_data.append(year)
            one_data.append(one_row[year].values[0])
            all_data.append(one_data)
all_data = pd.DataFrame(all_data)
all_data.columns = ['Area','Item','Year','Value']
all_data
```

```python
data = all_data
all_ = pd.DataFrame()
for i in data['Item'].unique():
    one_data = data[data['Item']==i]
    one_data.drop('Item',axis=1,inplace=True)
    one_data.columns = ['Area','Year',i]
    one_data.set_index(['Area','Year'],inplace=True)
    display(one_data)
    all_ = pd.concat([all_,one_data],axis=1)
all_.to_csv('政治指标.csv')
```
