```python
import tushare as ts
import pandas as pd 
import numpy as np 
ts.set_token('7c9f5442e7afbee1c52d71c8f3cc1931af510a048a85580a40530639')
pro = ts.pro_api()
```

```python
ticker = pd.read_excel('消费股代号.xlsx')
ticker_list = []
for row in ticker.values:
    ticker_new = row[0]
    ticker_new = str(ticker_new).zfill(6)
    ticker_list.append(ticker_new)
print(ticker_list)
```

```python
new_tickers = []
for i in ticker_list:
    if i[0] == '0' or i[0] == '3':
        i = i + '.SZ'
    if i[0] == '6':
        i = i + '.SH'
    new_tickers.append(i)   # 601238.SH  002044.SZ
new_tickers
```

```python
result = pd.DataFrame()
for ticker in new_tickers:
    data = pro.daily(ts_code=ticker, start_date='20150101')
    data.sort_values(by='trade_date',inplace=True)
    data.set_index('trade_date',inplace = True)
    data.drop(['ts_code','open','high','low','pre_close','change','pct_chg','vol','amount'],axis=1,inplace=True)
    data.columns = [ticker]
    result = pd.concat([result, data], axis=1, join='outer')
result.reset_index(inplace=True)
result.sort_values(by='index',inplace=True)
result.set_index('index',inplace=True)
result
```

```python
result.to_csv('消费者股指orginal.csv')
```
