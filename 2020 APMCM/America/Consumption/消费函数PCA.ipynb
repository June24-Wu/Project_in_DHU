```python
import tushare as ts
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('消费者股票+CPI.csv')
data.set_index('trade_date',inplace=True)
corr = data.corr()
# 丢掉PM烟草公司的影响
data.drop('PM',axis=1,inplace=True)
data

```

```python

corr
```

```python
from sklearn.feature_selection import VarianceThreshold  # sklearn 方差过滤API
from sklearn.decomposition import PCA  # 主成分分析
from sklearn.impute import SimpleImputer # 处理缺失值


var = VarianceThreshold(threshold=1)  # threshold 是门槛
var_data = var.fit_transform(data)
var_data = pd.DataFrame(var_data)
var_data.index = data.index
var_data  # finally drop the CPI 


# fill the missing value 
imputer = SimpleImputer(missing_values= np.NaN , strategy= 'mean')
var_data = imputer.fit_transform(var_data)
var_data = pd.DataFrame(var_data)
var_data.index = data.index
var_data

# pca 
pca = PCA(n_components=0.9)  # 一般 n_components 为0.9-0.95 左右
pca_data = pca.fit_transform(var_data)
pca_data = pd.DataFrame(pca_data)
pca_data.index = data.index 
pca_data



```

```python
pca_data.to_csv('C.csv')
```
