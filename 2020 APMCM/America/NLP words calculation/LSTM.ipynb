```python
import pandas as pd 
import keras.preprocessing.text as tt
import numpy as np 
from sklearn.model_selection import train_test_split
import keras.preprocessing.sequence as sequence
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('nlp_data.csv') # H: 1  , NA:2
data
```

```python
X = data['content'].tolist()
tokenizer = tt.Tokenizer(  nb_words=2000, 
                        filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
                        lower=True,split=' '
                     )
tokenizer.fit_on_texts(X)   # 转化为矩阵
X = tokenizer.texts_to_sequences(X)  # 转化为序列
X = sequence.pad_sequences(X)   # 填充长短不一的矩阵


Y = np.array(data['label'])
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size = 0.2)
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import LSTM , Dense , Embedding , Dropout,Flatten
import keras

embed_dim = 128
lstm_out = 256
batch_size = 64
epochs = 10
model = Sequential([
        Embedding(input_dim=30000, output_dim=32,input_length=x_train.shape[1]),
        LSTM(32, return_sequences=True),
        LSTM(1, activation='sigmoid', return_sequences=False)
    ])
model.compile(optimizer= keras.optimizers.Adam(),
                 loss=keras.losses.BinaryCrossentropy(),
                 metrics=['accuracy'])
model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.1)
```

```python
test_result = model.evaluate(x_test, y_test)
test_result
```
