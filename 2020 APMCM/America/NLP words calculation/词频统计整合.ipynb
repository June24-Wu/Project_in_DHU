```python
import pandas as pd 
import docx
import operator
```

```python
#获取文档对象
file=docx.Document("特朗普政治纲领.docx")
text = ''
for para in file.paragraphs:
    text += para.text
    text += ' '
print(text)
speech = text.lower().split()

# 利用字典进行处理
dic = {}
for word in speech:
    if word not in dic:
        dic[word] = 1
    else:
        dic[word] = dic[word] + 1
ci_pin = sorted(dic.items(),key=operator.itemgetter(1),reverse=True)
trump_debate = pd.DataFrame(ci_pin)
trump_debate.columns = ['word','frequency']
trump_debate.set_index('word',inplace=True)
trump_debate.to_csv('trump政治纲要统计.csv')
trump_debate
```

```python
with open("us_election_2020_trump_town_hall.txt", "rb") as f:  # 打开文件
    words = f.read()  # 读取文件
words = str(words)
words
import re
find = re.compile(r'Trump:(.*?)Savannah',re.S)  # 价格
def cleaner(word):
    word = re.sub(r'\\n', '', word)
    word = re.sub(r'\\n\\', '', word)
    word = re.sub(r'\\xe2\\x80\\x99', r"'" , word)
    word = re.sub(r'\\xe2\\x80\\x9d', r"'" , word)
    word = re.sub(r'\\xe2\\x80\\x9', r"'" , word)
    word = re.sub(r'\\xe2\\x80\\xa6', r"'" , word)
    return word
words = cleaner(words)
words = re.findall(find,words)
words_final = ""
for word in words:
    words_final += word
    words_final += ' '
    
speech = words_final.lower().split()

# 利用字典进行处理
dic = {}
for word in speech:
    if word not in dic:
        dic[word] = 1
    else:
        dic[word] = dic[word] + 1
ci_pin = sorted(dic.items(),key=operator.itemgetter(1),reverse=True)
trump_debate = pd.DataFrame(ci_pin)
trump_debate.columns = ['word','frequency']
trump_debate.set_index('word',inplace=True)
trump_debate.to_csv('trump_debate统计.csv')
trump_debate
```
