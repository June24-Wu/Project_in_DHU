```python
import pandas as pd
import tushare as ts
import numpy as np
from datetime import datetime
from datetime import timedelta
import random
```

```python
data = pd.read_csv('美国.csv')
data
```

```python

year_list = data['Year'].drop_duplicates()
season_list = data['Season'].drop_duplicates()
columns_value = [ i for i in data.columns]
columns_value.append('GDP_per_day')
```

```python
final_list = []
for year in year_list:
    for season in season_list:
        season_data = data[(data['Year']==year) & (data['Season']==season)]
        sum_hs300 = sum(season_data['Close'])
        for index , row in season_data.iterrows():
            sound = random.uniform(0.97,1.03)
            daily_gdp = (row['Close']/sum_hs300) * row['GDP Season'] * sound
            row = row.tolist()
            row.append(daily_gdp)
            final_list.append(row)
final_list = pd.DataFrame(final_list)
final_list.columns = columns_value
final_list
```

```python
final_list.to_csv('model data final.csv')
```
